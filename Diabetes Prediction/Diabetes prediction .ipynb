```julia
y = train["Outcome"]
```

```julia
X=select!(train,Not(:Outcome))
```

```julia
X= convert(Array,X)
```

```julia
y = convert(Array,y)
```

```julia
@sk_import model_selection: train_test_split
```

```julia
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```julia
using ScikitLearn
```

```julia
@sk_import tree:DecisionTreeClassifier
```

```julia
tree= DecisionTreeClassifier()
```

```julia
fit!(tree,X_train,y_train)
```

```julia
tree_pred = predict(tree,X_test)
```

```julia
@sk_import metrics:accuracy_score
```

```julia
acc= accuracy_score(tree_pred,y_test)
```

```julia
@sk_import ensemble : RandomForestClassifier
```

```julia
random =RandomForestClassifier()
```

```julia
fit!(random,X_train,y_train)
```

```julia
random_pred = predict(random,X_test)
```

```julia
random_acc = accuracy_score(random_pred,y_test)
```
